```
import pandas as pd
```

```
a = [1123, 1223, 1233, 1234, 2134, 1243, 4321, 4231, 4312, 3214]
b = ['julio', 'carla', 'manu', 'jo達o', 'bento', 'fernando', 'fernanda', 'bruno', 'theo', 'theodoro']
c = [40, 39, 7, 11, 1, 41, 23, 53, 23, 20]
d = ['aaa', 'bbb', 'ccc', 'ddd', 'eee', 'fff', 'ggg', 'hhh', 'iii', 'jjj']

dados = {'id':a, 
      'nome':b, 
      'idade':c, 
      'apelido':d}
```

```
x = pd.DataFrame(dados)
```

```
x
```

```
a1 = [1123, 1123, 1234, 2134, 2134]
b1 = ['julio', 'julio', 'jo達o', 'bento', 'bento']
c1 = [40, 40, 11, 1, 1]
d1 = [20, 10, 30, 5, 6]

dados2 = {'id':a1, 
      'idade':c1, 
      'pagamentos':d1}
```

```
y = pd.DataFrame(dados2)
y
```

```
resultado_inner = pd.merge(x, y,on=['id'], how='inner')
resultado_inner
```

```
resultado_left = pd.merge(x, y,on=['id'], how='left')
resultado_left
```

```
resultado_full = pd.concat([x, y],join='outer')
resultado_full
```

```
resultado_right = pd.merge(x, y,on=['id'], how='right')
resultado_right
```

```
resultado_right.groupby(['id','nome'])['pagamentos'].mean()
```

```
resultado_right.groupby(['id', 'nome']).sum()
```
